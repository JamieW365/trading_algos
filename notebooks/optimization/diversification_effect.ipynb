# The Portfolio Diversification Effect

```python
import pandas as pd
from matplotlib import pyplot as plt

from trading_algos import optimization as tao
from trading_algos import utils as tau
from trading_algos import datasets as tad

%load_ext autoreload
%autoreload 2
```

```python
sp_tickers = tad.get_sp500_tickers(get_latest=True)

df_ewps = pd.DataFrame()
df_ewps[['n_assets', 'i_run', 'tickers', 'risk', 'return', 'sharpe']] = None

ewp_n_assets = []
ewp_run_num = []
ewp_tickers = []
ewp_risk = []
ewp_return = []
ewp_sharpe = []

for n in range(1, 20):
    for i in range(50):
        tickers    = tao.select_n(n, sp_tickers)
        df_stocks  = tad.load_data(tickers, start_date='2020-01-01')
        df_returns = tao.portfolio_returns(df_stocks, 'EWP')
        df_summary = tau.get_returns_summary(df_returns)

        ewp_n_assets.append(n)
        ewp_run_num.append(i)
        ewp_tickers.append(tickers)
        ewp_risk.append(df_summary.loc['EWP', 'Risk'])
        ewp_return.append(df_summary.loc['EWP', 'Return'])
        ewp_sharpe.append(df_summary.loc['EWP', 'Sharpe'])

        print(df_summary.head())

df_ewps = pd.DataFrame({'N Assets': ewp_n_assets,
                        'Simulation': ewp_run_num,
                        'EWP Stocks': ewp_tickers,
                        'Risk': ewp_risk,
                        'Return': ewp_return,
                        'Sharpe': ewp_sharpe})
df_ewps.set_index(['N Assets', 'Simulation'], inplace=True)
```
